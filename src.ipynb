```python
count = 0
correct = 0
err = 0
test_playlist = {}
for playlist in test_data['playlists']:
    if playlist['num_samples'] == 10 and "name" in playlist:
        tracks = playlist['tracks']
        right = True
        for t in tracks:
            track_uri = t['track_uri'].split(':')[2]
            if not track_uri in track_list:
                right = False
                err = err + 1
        if right:
            track_l = []
            for t in tracks: 
                track_l.append(t['track_uri'].split(':')[2])
            correct = correct + 1
            test_playlist[playlist['name']] = track_l
        count = count + 1
print(count)
print(correct)
print(err)
```

```python
playlist_sim = np.append(playlist['playlist'].to_numpy().reshape(-1,1),playlist_sim, axis = 1)
```

```python
playlist_sim = playlist_sim[playlist_sim[:,1].argsort()[::-1]]
display(playlist_sim)
```

```python
track_score = np.zeros(len(track))
print(track_score.shape)
for p in playlist_sim:
    name = p[0]
    score = p[1]
    playlist_track_list = table.loc[table['playlist'] == name]['track_uri']
    vect = np.array(track['track_uri'].isin(playlist_track_list)*1)
    vect = vect*score
    track_score = track_score + vect
```

```python
target_track = np.append(track_score.reshape(-1,1),np.array(track['track_uri']).reshape(-1,1), axis = 1)
target_track = target_track[target_track[:,0].argsort()[::-1]]
```

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm') 
```

```python
a = nlp("morning")
b = nlp("mornign")
c = nlp("dawn")
d = nlp("night")
print(a.similarity(b))
print(a.similarity(c))
print(a.similarity(d))
```

```python
print("Enter two space-separated words") 
words = input() 
spacy.load('en')
tokens = nlp(words) 
  
for token in tokens: 
    # Printing the following attributes of each token. 
    # text: the word string, has_vector: if it contains 
    # a vector representation in the model,  
    # vector_norm: the algebraic norm of the vector, 
    # is_oov: if the word is out of vocabulary. 
    print(token.text, token.has_vector, token.vector_norm, token.is_oov) 
  
token1, token2 = tokens[0], tokens[1] 
  
print("Similarity:", token1.similarity(token2)) 
```
